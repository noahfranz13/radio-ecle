```python
import numpy as np
import pandas as pd
from astropy import units as u, constants as c
from astropy.cosmology import Planck18 as cosmo
```

```python
def condonSFR(lum):
    return 5.5 * (lum / 4.6e21) 
```

```python
df = pd.DataFrame(dict(
    names = ["SDSSJ0938", "SDSSJ0748", "SDSSJ1241-1", "SDSSJ1241-2"],
    Lband_flux = np.array([2752.3, 215.6, 4281.6, 1451.771317]) * u.uJy,
    z = [0.101, 0.062, 0.042, 0.042]
))

df["Sband_lum"] = (4 * np.pi * cosmo.luminosity_distance(df.z.values)**2 * df.Lband_flux.values*u.uJy).to(u.W/u.Hz).value

df["sfr"] = condonSFR(df.Sband_lum.values)
```

```python
def kennicut_halpha_sfr(ha_lum):
    """From Kennicutt 1998"""
    return 7.9e-42 * ha_lum / (u.erg/u.s) * u.Msun/u.yr

z = 0.062
flux = 213 * 1e-17 * u.erg/u.cm**2/u.s

lum = (4*np.pi*flux*cosmo.luminosity_distance(z)**2).to(u.erg/u.s)
kennicut_halpha_sfr(lum)
```
