```python
import pandas as pd

alldata = pd.read_csv("ecle-radio-photometry.csv")
data = alldata[(alldata.source.str.contains("-") + alldata.source.str.contains("_"))*(alldata.use_this)]

forlatex = {
    "ECLE":data.name.str.replace("_","\,").tolist(),
    "MJD":data.date.tolist(),
    "Frequency (GHz)": data.filter_eff.tolist(),
    "Flux Density ($\mu$Jy)": data.apply(lambda row : f"${row.flux:.2f} \pm {row.flux_err:.2f}$" if not row.upperlimit else f"$<{row.flux:.2f}$", axis=1).tolist(),
    "Telescope": ["GMRT" if "_" in s  else "VLA" for s in data.source],
    "Project Code": data.source.tolist()
}
```

```python
print(
    pd.DataFrame(forlatex).sort_values(
        ["ECLE", "Project Code", "MJD", "Frequency (GHz)"]
    ).to_latex(
        index=False,
        float_format="{:0.2f}".format,
        longtable=True
    )
)
```
